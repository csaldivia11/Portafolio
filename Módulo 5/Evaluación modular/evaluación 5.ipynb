```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch, ConnectionPatch
import numpy as np

# Crear figura
fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')

# Colores
color_fact = '#FFE6E6'  # Rosa claro para tabla de hechos
color_dim = '#E6F3FF'   # Azul claro para dimensiones
color_border = '#333333'

# Función para crear cajas redondeadas
def create_box(ax, x, y, width, height, text, color, fontsize=10):
    box = FancyBboxPatch((x, y), width, height,
                         boxstyle="round,pad=0.1",
                         facecolor=color,
                         edgecolor=color_border,
                         linewidth=2)
    ax.add_patch(box)
    ax.text(x + width/2, y + height/2, text, 
            ha='center', va='center', fontsize=fontsize, fontweight='bold')

# Función para crear conexiones
def create_connection(ax, start_x, start_y, end_x, end_y):
    line = ConnectionPatch((start_x, start_y), (end_x, end_y), "data", "data",
                          arrowstyle="->", shrinkA=5, shrinkB=5, 
                          mutation_scale=20, fc=color_border, lw=2)
    ax.add_patch(line)

# TABLA DE HECHOS (Centro)
fact_text = """FACT_VENTAS
(Tabla de Hechos)

🔑 id_venta (PK)
🔗 id_producto (FK)
🔗 id_sucursal (FK)
🔗 fecha (FK)

📊 MÉTRICAS:
💰 cantidad
💵 monto
💸 ingresos_total

📈 DERIVADOS:
📅 año, mes, trimestre
🏷️ categoria_venta"""

create_box(ax, 3.5, 4, 3, 3, fact_text, color_fact, 9)

# DIMENSIÓN PRODUCTOS (Izquierda)
prod_text = """DIM_PRODUCTOS

🔑 id_producto (PK)
📦 nombre_producto
🏷️ categoria
🏭 proveedor"""

create_box(ax, 0.5, 4.5, 2.5, 2, prod_text, color_dim, 9)

# DIMENSIÓN SUCURSALES (Derecha)
suc_text = """DIM_SUCURSALES

🔑 id_sucursal (PK)
📍 ubicacion
(expandible)"""

create_box(ax, 7, 4.5, 2.5, 2, suc_text, color_dim, 9)

# DIMENSIÓN TIEMPO (Arriba)
time_text = """DIM_TIEMPO

🔑 fecha (PK)
📅 año
📅 mes
📅 trimestre"""

create_box(ax, 3.5, 8, 3, 1.5, time_text, color_dim, 9)

# DIMENSIÓN CLIENTES (Abajo)
client_text = """DIM_CLIENTES

🔑 id_cliente (PK)
📝 nombre
🔢 edad
📍 ubicacion
🏷️ categoria"""

create_box(ax, 3.5, 1, 3, 2, client_text, color_dim, 9)

# CONEXIONES
# Productos a Hechos
create_connection(ax, 3, 5.5, 3.5, 5.5)

# Sucursales a Hechos
create_connection(ax, 7, 5.5, 6.5, 5.5)

# Tiempo a Hechos
create_connection(ax, 5, 8, 5, 7)

# Clientes a Hechos (línea punteada para futuro)
ax.plot([5, 5], [3, 4], 'k--', linewidth=2, alpha=0.5)
ax.text(5.2, 3.5, 'Futuro', fontsize=8, style='italic')

# Título
ax.text(5, 9.5, 'MODELO MULTIDIMENSIONAL - TECHMART', 
         ha='center', va='center', fontsize=16, fontweight='bold')

# Subtítulo
ax.text(5, 9.2, 'Esquema Estrella (Star Schema)', 
         ha='center', va='center', fontsize=12, style='italic')

# Leyenda
legend_text = """LEYENDA:
🔑 = Clave Primaria (PK)
🔗 = Clave Foránea (FK)
📊 = Métricas/Measures
📈 = Atributos Derivados
—— = Relación Activa
- - - = Relación Futura"""

ax.text(0.5, 1.5, legend_text, fontsize=8, va='top',
        bbox=dict(boxstyle="round,pad=0.3", facecolor='#F0F0F0', alpha=0.8))

plt.tight_layout()
plt.savefig('diagrama_multidimensional_techmart.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Diagrama guardado como 'diagrama_multidimensional_techmart.png'")
print("📊 Resolución: 300 DPI para calidad profesional")
print("🎨 Colores: Rosa para hechos, Azul para dimensiones")
```
